```python
# type of kurtosis
if smoke_kurt > 3:
    print('\n Smoking Prevalence is Leptokurtic.')
elif smoke_kurt < 3:
    print('\n Smoking Prevalence is Platykurtic.')
else:
    print('\n Smoking Prevalence is Mesokurtic.')

if drug_kurt > 3:
    print('\n Drug Experimentation is Leptokurtic.')
elif drug_kurt < 3:
    print('\n Drug Experimentation is Platykurtic.')
else:
    print('\n Drug Experimentation is Mesokurtic.')

if peer_kurt > 3:
    print('\n Peer Influence is Leptokurtic.')
elif peer_kurt < 3:
    print('\n Peer Influence is Platykurtic.')
else:
    print('\n Peer Influence is Mesokurtic.')
```

# group data make Freq distribution table

```python
numeric_columns = my_data.select_dtypes(include=[np.number]).columns
num_bins = 10
results = {}
for column in numeric_columns:
    print('\nProcessing column: ', column)
    my_data[column + '_bins'] = pd.cut(my_data[column], bins=num_bins)
    freq_table = my_data[column + '_bins'].value_counts().sort_index()
    bin_midpoints = [(interval.left + interval.right) / 2 for interval in freq_table.index]
    frequencies = freq_table.values

    mean = sum(mid * freq for mid, freq in zip(bin_midpoints, frequencies)) / sum(frequencies)
    variance = sum(freq * (mid - mean)**2 for mid, freq in zip(bin_midpoints, frequencies)) / sum(frequencies)
    std_dev = np.sqrt(variance)

    # Quartiles and IQR
    quartiles = np.percentile(my_data[column].dropna(), [25, 50, 75])
    iqr = quartiles[2] - quartiles[0]

    # Deciles
    deciles = np.percentile(my_data[column].dropna(), np.arange(10, 100, 10))

    # Moments (1st, 2nd, 3rd, 4th)
    moments = {
        '1st Moment (Mean)': mean,
        '2nd Moment (Variance)': variance,
        '3rd Moment (Skewness)': sum(freq * (mid - mean)**3 for mid, freq in zip(bin_midpoints, frequencies)) / (sum(frequencies) * std_dev**3),
        '4th Moment (Kurtosis)': sum(freq * (mid - mean)**4 for mid, freq in zip(bin_midpoints, frequencies)) / (sum(frequencies) * std_dev**4) - 3,
    }

    # Coefficient of Variation
    coef_var = (std_dev / mean) * 100

    #  results
    results[column] = {
        'Frequency Table': freq_table,
        'Mean': mean,
        'Variance': variance,
        'Standard Deviation': std_dev,
        'Coefficient of Variation': coef_var,
        'Quartiles': quartiles,
        'IQR': iqr,
        'Deciles': deciles,
        'Moments': moments
    }

# Display results
for column, stats in results.items():
    print('\nColumn: ', column)
    print('\nFrequency Table:')
    print(stats['Frequency Table'])
    print('\nMean: ', stats['Mean'])
    print('Variance: ', stats['Variance'])
    print('Standard Deviation: ', stats['Standard Deviation'])
    print('Coefficient of Variation: ', stats['Coefficient of Variation'], '%')
    print('Quartiles (Q1, Q2, Q3): ', stats['Quartiles'])
    print('Interquartile Range (IQR): ', stats['IQR'])
    print('Deciles: ', stats['Deciles'])
    print('Moments:')
    for moment_name, moment_value in stats['Moments'].items():
        print('  ', moment_name, ': ', moment_value)
```
